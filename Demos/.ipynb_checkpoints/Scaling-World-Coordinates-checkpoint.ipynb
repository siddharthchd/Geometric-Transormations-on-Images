```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from skimage.color import rgb2gray
from scipy import signal
import math
```

```python
# Read the image and be sure it is grayscale
im_path = 'cameraman.png'
im = io.imread(im_path)
im = rgb2gray(im)

# Ensure the image is a float in the range [0-1]
im = ((im - np.min(im)) * (1/(np.max(im) - np.min(im)) *1.0)).astype('float')

# The columns correpsond to x
# The rows correspond to y 
[rows, cols] = im.shape

#################################################################### 
# Represent the image with "world coordinates" between -1 and 1
# See equation below to rescale from [a,b] to [c,d]
####################################################################

# First for x (columns)
a1 = 0.0
b1 = cols
c1 = -1.0
d1 = 1.0

# And then for y (rows)
a2 = 0.0
b2 = rows
c2 = -1.0
d2 = 1.0

# These are the original pixel coordinates
x = np.linspace(0, cols, cols)
y = np.linspace(0, rows, rows)

# These are the *world coordinates* of the image, which we center at the origin (0,0)
# i.e. the center of the image is the origin instead of the upper left
world_x = c1+((d1-c1)/(b1-a1))*(x-a1)
world_y = c2+((d2-c2)/(b2-a2))*(y-a2)

# Some print statements for a sanity check

print("Original Image Dimensions:")
print("  Cols (x) = ", str(cols))
print("  Rows (y) = ", str(rows))

print("\nWorld Coordinate Dimensions")
print('  Cols (x) = ' + str(len(world_x)))
print('  Rows (y) = ' + str(len(world_y)))
```

```python
# The scaling matrix for Sx = 2.0 and Sy = 3.0
S = np.array([[1/2, 0, 0], [0, 1/3, 0], [0,0,1]])
Sinv = np.linalg.inv(S)

# Compute forward transformation to determine size of canvas

# Top left corner
top_left = Sinv @ np.array([world_x[0], world_y[0], 1.0], float)
top_left = top_left / top_left[2]

# Bottom left corner
bottom_left = Sinv @ np.array([world_x[0], world_y[rows-1], 1.0], float)
bottom_left = bottom_left / bottom_left[2]

# Top right corner
top_right = Sinv @ np.array([world_x[cols-1], world_y[0], 1.0], float)
top_right = top_right / top_right[2]

# Bottom right corner
bottom_right = Sinv @ np.array([world_x[cols-1], world_y[rows-1], 1.0], float)
bottom_right = bottom_right / bottom_right[2]

# Bounding box for the canvas
min_x = np.min([top_left[0], bottom_left[0], top_right[0], bottom_right[0]])
max_x = np.max([top_left[0], bottom_left[0], top_right[0], bottom_right[0]])

min_y = np.min([top_left[1], bottom_left[1], top_right[1], bottom_right[1]])
max_y = np.max([top_left[1], bottom_left[1], top_right[1], bottom_right[1]])

# Calculate the scale factor
# Even though we know these it is good practice to make the code generic
scale_factor_x = (max_x-min_x)/(world_x[len(world_x)-1]-world_x[0])
scale_factor_y = (max_y-min_y)/(world_y[len(world_y)-1]-world_y[0])

# Create the *world coordinates* for the canvas
out_world_x = np.linspace(min_x, max_x, int(scale_factor_x*cols))
out_world_y = np.linspace(min_y, max_y, int(scale_factor_y*rows))

# Create the blank output image canvas
# Recall it is rows (y) x columns (x)
out_im = np.zeros((len(out_world_y), len(out_world_x)))

# Some print statements for a sanity check

print("Desired Canvas Dimensions:")
print("  Cols (x) = ", str(2*cols))
print("  Rows (y) = ", str(3*rows))

print("\nOut World Coordinate Dimensions")
print('  Cols (x) = ' + str(len(out_world_x)))
print('  Rows (y) = ' + str(len(out_world_y)))

print("\nOut Canvas Dimensions")
print('  Cols (x) = ' + str(out_im.shape[1]))
print('  Rows (y) = ' + str(out_im.shape[0]))
```

```python
# Loop over the rows (y coordinates)
for cur_row in range(0, len(out_world_y)):
    # Loop over the cols (x coordinates)
    for cur_col in range(0, len(out_world_x)):
        
        # Transform the *world coordinate* in the canvas and see where it lands in the original image
        
        # 1) Get the *world coordinate* in the canvas
        out_im_world_coordinate = np.array([out_world_x[cur_col], out_world_y[cur_row], 1.0])
        
        # 2) Transform it via matrix multiplication (@ is python matrix multiplication)
        transformed_pt = S @ out_im_world_coordinate
        transformed_pt = transformed_pt/transformed_pt[2]
        
        # 3) Now we have transformed world coordinates which land in the original image.
        #    We must find the corresponding INDEX in the original image (NN interpolation)
        lookup_row = int(np.abs(world_y - transformed_pt[1]).argmin())   # Row is y
        lookup_col = int(np.abs(world_x - transformed_pt[0]).argmin())   # Column is x
        
        # Make sure we don't go off of the original image
        if (lookup_col < 0 or lookup_col >= cols or lookup_row < 0 or lookup_row >= rows):
            continue
        
        out_im[cur_row,cur_col] = im[lookup_row, lookup_col]
        
        
# Show the results
fig1 = plt.figure(figsize = (7,7))
plt.subplot(1,2,1)
plt.imshow(im, cmap='gray', aspect='equal')
plt.subplot(1,2,2)
plt.imshow(out_im, cmap='gray', aspect='equal')

# Save the image in the working directory
from PIL import Image
import numpy as np

#Rescale to 0-255 and convert to uint8
rescaled = (255.0 / out_im.max() * (out_im - out_im.min())).astype(np.uint8)

save_im = Image.fromarray(rescaled)
save_im.save('rescaled.png')
```
